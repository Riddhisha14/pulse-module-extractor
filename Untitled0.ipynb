```python
url = "https://support.spotify.com/"
response = requests.get(url)

print("Spotify help website opened")
```

```python
soup = BeautifulSoup(response.text, "html.parser")

headings = soup.find_all(["h1", "h2", "h3"])

for h in headings[:20]:
    print(h.text.strip())
```
